```python
std=StandardScaler()
std_indicators=indicators.drop(['Date','Recession1','Recession6','Recession12'],axis=1)                      #change later

std.fit(std_indicators)
standardized_features=std.transform(std_indicators)

finalInd=pd.DataFrame(data=standardized_features,columns=std_indicators.columns)
finalInd.insert(loc=0,column="Date", value=indicators['Date'].values)
finalInd.insert(loc=1,column='Recession1', value=indicators['Recession1'].values)
finalInd.insert(loc=2,column='Recession6', value=indicators['Recession6'].values)
finalInd.insert(loc=3,column='Recession12', value=indicators['Recession12'].values)


finalInd.head()

```
